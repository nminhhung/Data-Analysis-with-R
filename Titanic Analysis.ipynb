```r
getmode = function(v){
  uniqv = unique(v)
  uniqv[which.max(tabulate(match(v, uniqv)))]
}
```

```r
desctiptive_statitic = function(x){
  print(paste("Mean:", mean(x)))
  print(paste("Median: ", median(x)))
  print(paste("Mode: ", getmode(x)))
  print(paste("Max: ", range(x)[2]))
  print(paste("Min: ", range(x)[1]))
  print(paste("Range: ", range(x)[2] - range(x)[1]))
}
```

```r
desctiptive_statitic(titanic_sub$Age)
```

```r
desctiptive_statitic(titanic_sub$Fare)
```

```r
nums <- unlist(lapply(titanic_sub, is.numeric), use.names = FALSE)  
```

```r
sapply(titanic_sub[, nums], function(x) c(var=var(x), sd=sd(x), skewness = skewness(x), kurtosis = kurtosis(x)))
```

NX: 

`skewness` khong co cot nao = 0. 
  - `Pclass` < 0: phan phoi lech trai.
  - Cac cot con lai > 0: phan phoi lech phai.

`kurtosis`khong co cot nao = 3.
  - `Survived` & `Pclass` < 3: thap hon phan phoi chuan.
  - Cac cot con lai > 3: cao hon phan phoi chuan.

NX: `Age` va `Fare` khong co moi quan he tuyen tinh voi nhau.

```r
# covariance & correlation
cov_cor = function(x, y){
  print(paste("Covariance: ", cov(x, y)))
  print(paste("Correlation: ", cor(x, y)))
}
```

```r
cov_cor(titanic_sub$Age, titanic_sub$Fare)
```

NX: `covariance`: 73.8, chung to 2 bien co quan he dong bien.
`correlation`: 0.09, rat be, chung to 2 bien khong co quan he chat che voi nhau.
