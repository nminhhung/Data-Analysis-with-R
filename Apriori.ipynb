```r
tr1 = read.transactions("/content/ItemList.csv", format = "basket", sep = ",", header=T)
inspect(tr1)
itemLabels(tr1)
```

### Apriori 

```r
# min sup = .00001, min confidence = .0002
rules1 = apriori(tr1, parameter = list(supp = .001, conf = .02, target="rules"))
```

```r
print(length(rules1))
```

```r
summary(rules1)
```

```r
# inspect function
inspect(sort(rules1, by = "support", decreasing = T))
```

```r
# filter
filter_rules1 = subset(rules1, subset = lift > 1.2)
summary(filter_rules1)
```

```r
inspect(filter_rules1)
```

```r
# filter top 15
arules::itemFrequencyPlot(tr1, topN = 15,
                          col = brewer.pal(5, "Pastel2"),
                          main = "Relative item Frequency Plot",
                          type = "relative",
                          ylab = "item frequency (relative)"
                          )
```
