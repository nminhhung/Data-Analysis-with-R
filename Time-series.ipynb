```r
ts.df = ts(df$price, frequency = 12, start = c(2004,1,1))
```

```r
ts.df
```

```r
plot.ts(ts.df, main = "Rice price from 2004 to 2022 in VN",
          ylab = "price",
          xlab = "time",
          col = "red")
```

```r
td = decompose(ts.df)
```

```r
plot(td)
```

```r
plot(td$seasonal + td$trend + td$random)
```

```r
HWSE.forecast = HoltWinters(log(ts.df))
```

```r
HWSE.forecast
```

```r
plot(HWSE.forecast)
```
